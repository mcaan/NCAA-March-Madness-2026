### Calculating Z-scores

##### Box Score metrics are standardized here as z-scores, allowing for direct comparison of predictor importance
##### This is largely done to put variables on the same scale, making relative importance clear
> ##### Shots Attempted is likely to be on a larger scale than Steals or Blocks
> ##### EFG is a percentage, while other measures are integers

```python
### Offensive Efficiency

# Winning/Losing Teams' Effective Field Goal Percentage (efg)
RegularDetailedResults['WEFG'] = (RegularDetailedResults['WFGM'] + 
    (1.5*RegularDetailedResults['WFGM3']))/(RegularDetailedResults['WFGA'] + 
                                         RegularDetailedResults['WFGA3'])
RegularDetailedResults['LEFG'] = (RegularDetailedResults['LFGM'] + 
    (1.5*RegularDetailedResults['LFGM3']))/(RegularDetailedResults['LFGA'] + 
                                         RegularDetailedResults['LFGA3'])
# Winning/Losing Teams' Shots Attempted (sa)
# sa = fga + fga3
RegularDetailedResults['WSA'] = (RegularDetailedResults['WFGA'] + 
                                 (RegularDetailedResults['WFGA3']))
RegularDetailedResults['LSA'] = (RegularDetailedResults['LFGA'] + 
                                 (RegularDetailedResults['LFGA3']))

display(RegularDetailedResults.head())
```

##### Effective Field Goal Perfentage (eFG%) adjusts traditional field goal percentage to account for the higher value of 3-point shots
##### efg = (field goals made + (0.5 * 3pt shots made)) / (field goals attempted + 3pt shots attemtped)

```python
# Select relevant columns and rename for winning/losing teams; merge
e1 = RegularDetailedResults[['Season', 'WTeamID', 'WEFG', 'WSA']].rename(
    columns = {'Season':'season', 'WTeamID':'team', 'WEFG':'EFG', 'WSA':'shot_attempts'})
e2 = RegularDetailedResults[['Season', 'LTeamID', 'LEFG', 'LSA']].rename(
    columns = {'Season':'season', 'LTeamID':'team', 'LEFG':'EFG', 'LSA':'shot_attempts'})
e3 = pd.concat([e1,e2])

# Add conference column to e3
# Rename TeamConf df columns for merge w/ Teams df, merge w/ e3
TeamConf = TeamConf.rename(columns = {'season':'season', 'team_id':'Team_Id', 'conference':'Conf_ID'})
Team_with_Conf = TeamConf.merge(Teams, how='right', on='Team_Id')[[
    'season', 'Team_Id', 'Conf_ID']].rename(columns={'season':'season', 'Team_Id':'team', 'Conf_ID':'conf'})
# Merge Conf df with main df
e4 = e3.merge(Team_with_Conf, how='left', on = ['team', 'season'])

# Get mean and std dev of efficiency variables for each team
# This is the mean/std for each team by season and conference
e5 = e4.groupby(['season', 'team', 'conf']).mean().dropna().reset_index()
e5.columns = ['season', 'team', 'conf', 'efg_mean', 'sa_mean']

# Get season league-wide efficiency mean/std for z-score
e6 = e5.groupby('season').agg({'efg_mean':['mean', 'std'], 'sa_mean':['mean', 'std']}).dropna().reset_index()
e6.columns = ['_'.join(col).strip() for col in e6.columns.values] 
e6.columns = ['season', 'league_efg_mean', 'league_efg_std', 'league_sa_mean', 'league_sa_std']

# Merge with team-level df
e7 = e6.merge(e5, how='right', on='season')

# calculate z-scores for efg
e7['z.efg'] = (e7['efg_mean'] - e7['league_efg_mean']) / e7['league_efg_std']
# calculate zscore for sa (to interact the efficient field goal with number of shot attempts) 
# 1 for 1 shouldn't be the same as 10 for 10
e7['z.sa'] = (e7['sa_mean'] - e7['league_sa_mean']) / e7['league_sa_std']

# select meaningful variables
e7 = e7[['team', 'season', 'conf', 'z.efg', 'z.sa']]
display(e7.head(), e7.shape)
```

```python
### Z-score of average point differential of each team by conference by year over the regular season

# Create point difference variables for each team in each matchup
RegularDetailedResults['Wptdiff'] = RegularDetailedResults['WScore'] - RegularDetailedResults['LScore']
RegularDetailedResults['Lptdiff'] = RegularDetailedResults['LScore'] - RegularDetailedResults['WScore']

# Select relevant columns and rename for winning/losing teams; merge
p1 = RegularDetailedResults[['Season', 'WTeamID', 'Wptdiff']].rename(
    columns = {'Season':'season', 'WTeamID':'team', 'Wptdiff':'ptdiff'})
p2 = RegularDetailedResults[['Season', 'LTeamID', 'Lptdiff']].rename(
    columns = {'Season':'season', 'LTeamID':'team', 'Lptdiff':'ptdiff'})
p3 = pd.concat([p1,p2])

# Add conference column to p3
p4 = p3.merge(Team_with_Conf, how='left', on = ['team', 'season'])

# Get mean and std dev of efficiency variables for each team
# This is the mean/std for each team by season and conference
p5 = p4.groupby(['season', 'team', 'conf']).mean().dropna().reset_index()
p5.columns = ['season', 'team', 'conf', 'ptdiff_mean']

# calculate league mean and sd
# add conf
p6 = p5.groupby('season').agg({'ptdiff_mean':['mean', 'std']}).dropna().reset_index()
p6.columns = ['_'.join(col).strip() for col in p6.columns.values] 
p6.columns = ['season', 'league_ptdiff_mean', 'league_ptdiff_std']

# Merge with team-level df
p7 = p6.merge(p5, how='right', on='season')

# calculate z-scores for pt diff
p7['z.pt'] = (p7['ptdiff_mean'] - p7['league_ptdiff_mean']) / p7['league_ptdiff_std']

# select meaningful variables
p7 = p7[['team', 'season', 'conf', 'z.pt']]
display(p7.head(), p7.shape)
```

```python
### Caculate momentum heading into the post-season

# Calculate W-L record at day 100 for each team by season

# Identify teams in each matchup in every season
m1 = SampleSubmission['ID'].str.split('_', expand=True).iloc[:, ::-1]  # reverse the 3 columns
m1.columns = ['season', 'team1', 'team2']

# Combine with df for game-level stats
m2 = RegularCompactResults.merge(full_set, how='left', left_on=['Season', 'WTeamID'], right_on=['season', 'team']).drop(
    ['season', 'team'], axis=1)
m3= m2.merge(full_set, how='left', left_on=['Season', 'LTeamID'], right_on=['season', 'team']).drop(['season', 'team'], axis=1)

#
m4 = m3[['Season', 'WTeamID', 'DayNum']].rename(columns={'Season':'season', 'WTeamID':'team', 'DayNum':'day'}).assign(team1win = 1)
m5 = m3[['Season', 'LTeamID', 'DayNum']].rename(columns={'Season':'season', 'LTeamID':'team', 'DayNum':'day'}).assign(team1win = 0)
m6 = pd.concat([m5,m4])

#
m7 = m6.merge(Team_with_Conf, how='left', on=['team', 'season'])

#
m8 = m7[m7['day'] <= 100].groupby(['team', 'season', 'conf']).mean().dropna().reset_index().rename(
    columns = {'team1win':'winsbefore'}).sort_values(['season', 'winsbefore'], ascending=[True, False])

#
m8['rank'] = m8.groupby(['season','conf']).winsbefore.rank(method='dense', ascending=False).astype(int)
        
# Create categorical rankings based on ranking in conference
m8['top'] = (m8['rank'] <= 3).astype(int)
m8['mid'] = ((m8['rank'] >= 4) & (m8['rank'] <= 7)).astype(int)
m8['bot'] = (m8['rank'] >= 8).astype(int)    

# Calculate W-L record after day 100 for each team by season
m9 = m7[m7['day'] >= 101].groupby(['team', 'season', 'conf']).mean().dropna().reset_index().rename(
    columns = {'team1win':'winsafter'}).sort_values(['season', 'winsafter'], ascending=[True, False])

# Add winsafter to rankings df
m10 = m8.merge(m9, how='left', on=['team', 'season', 'conf']).drop(['day_x', 'day_y'], axis = 1)

# Normalize winsbefore and winsafter by season
m10['z.winsbefore'] = m10.groupby('season')['winsbefore'].transform(
    lambda x: (x - x.mean()) / x.std()
)

m10['z.winsafter'] = m10.groupby('season')['winsafter'].transform(
    lambda x: (x - x.mean()) / x.std()
)

m10 = m10.drop(['winsbefore','winsafter'], axis=1)

display(m10.head(), m10.shape)
```

##### Momentum is a created variable that looks at how many wins a team has going in to the postseason
> ##### This is calculated by the number of games won after day 100 of the season
##### In addition, teams are ranked at the day 100 mark to further look at whether momentum heading into the post season matters more for teams that were better-/worse- performing prior to that point
> ##### 'Top' = ranked 3rd or better within their conference
> ##### 'Mid' = ranked between 4th and 7th within their conference
> ##### 'Bot' = ranked 8th or worse within their conference

##### How much momentum a team has going in to the post season is a popular talking point among pundits leading up to March Madness
##### Winning late-season games may also be particularly important for 'bubble teams', or teams that don't automatically qualify for March Madness by winning their conference tournament, but still have a chance to make it via the selection committee

```python
### Clutch - how many close games (pt. diff <= 5 points)

# Start from p4 - already has ptdiff calculated
c1 = p4.copy()

# Create necessary variables for Clutch calculations
c1['close'] = c1['ptdiff'].between(-5, 5)
c1['team1win'] = c1['ptdiff'] > 0

clutch = (
    c1[c1['close']]
    .groupby(['team','season','conf'])['team1win']
    .mean()
    .rename('clutch_mean')
)

close_rate = (
    c1.groupby(['team','season','conf'])['close']
    .mean()
    .rename('close_mean')
)

c2 = pd.concat([clutch, close_rate], axis=1).reset_index()

# Normalize clutch_mean and close_mean by season
c2['z.clutch'] = c2.groupby('season')['clutch_mean'].transform(
    lambda x: (x - x.mean()) / x.std()
)

c2['z.close'] = c2.groupby('season')['close_mean'].transform(
    lambda x: (x - x.mean()) / x.std()
)

c2 = c2.drop(['clutch_mean','close_mean'], axis=1)

display(c2.head(), c2.shape)
```

##### Clutch is a metric created to calculate the number of close games a team has won
> ##### 'close' games are defined as a game that had a final score within 5 points
> ##### clutch wins are the number of close wins / number of close games (% of close games won)

##### Also calculated is the percentage of close games a team had, close_z
> ##### Defined as number of close games / total games, standardized

##### Teams that were able to win more close games may do better under pressure or may be more reliable in tough situations. There are different scenarios where this may occur: games that are close throughout the contest, where a team overcomes a large deficit, or where a team holds off a resurging opponent are just a few. These teams may perform better in high-stakes situations like March Madness

```python
### Grit - how many games with overtime won 

# Start from m3 - has necessary variables for calculations (OT)
g1 = m3[['Season', 'WTeamID', 'DayNum', 'NumOT']].assign(team1win = 1).rename(
    columns={'Season':'season', 'WTeamID':'team', 'DayNum':'day', 'NumOT':'ot'})
g2 = m3[['Season', 'LTeamID', 'DayNum', 'NumOT']].assign(team1win = 0).rename(
    columns={'Season':'season', 'LTeamID':'team', 'DayNum':'day', 'NumOT':'ot'})
g3 = pd.concat([g1,g2])

# Add conference data
g4 = g3.merge(Team_with_Conf, how='left', on=['team', 'season'])

# Create dummy, where =1 if game had an OT
g4['has_ot'] = (g4['ot'] > 0)

# Calculate Grit

grit = (
    g4[g4['has_ot']]  # only OT games
    .groupby(['team', 'season', 'conf'])['team1win']
    .mean()
    .rename('grit_mean')
)

ot_freq = (
    g4.groupby(['team', 'season', 'conf'])['has_ot']
    .mean()
    .rename('has_ot_mean')
)

#
g5 = pd.concat([grit, ot_freq], axis=1).reset_index()

# Add z-scores for grit_mean and has_ot_mean by season
g5['z.grit'] = g5.groupby('season')['grit_mean'].transform(
    lambda x: (x - x.mean()) / x.std()
)

g5['z.has_ot'] = g5.groupby('season')['has_ot_mean'].transform(
    lambda x: (x - x.mean()) / x.std()
)

g5 = g5.drop(['grit_mean','has_ot_mean'],axis=1)

display(g5.head(), g5.shape)
```

##### Grit is the last created metric, measuring the number of overtime games that a team has won
> ##### Grit_z is a team's proportion of OT wins in that season, standardized
> ##### Similarly, has_ot_z is the proportion of OT games that team had each year, standardized

```python
### Z-score of point totals of each team within each season across all teams

o1 = RegularCompactResults[['Season', 'WTeamID', 'WScore']].rename(
    columns={'Season':'season', 'WTeamID':'team', 'WScore':'score'})
o2 = RegularCompactResults[['Season', 'LTeamID', 'LScore']].rename(
    columns={'Season':'season', 'LTeamID':'team', 'LScore':'score'})
o3 = pd.concat([o1,o2])

# calculate team mean by conference and season
o4 = o3.merge(Team_with_Conf, how='left', on=['team', 'season'])
o5 = o4.groupby(['team', 'season', 'conf']).mean().rename(
    columns={'score':'score_mean'}).dropna().reset_index()

# calculate league mean and sd
# add conf
o6 = o5.groupby('season').agg({'score_mean':['mean', 'std']}).dropna().reset_index()
o6.columns = ['_'.join(col).strip() for col in o6.columns.values] 
o6.columns = ['season', 'league_score_mean', 'league_score_std']

# Merge with team-level df
o7 = o6.merge(o5, how='right', on='season')

# calculate z-scores for efg
o7['z.off'] = (o7['score_mean'] - o7['league_score_mean']) / o7['league_score_std']

# select meaningful variables
o7 = o7[['team', 'season', 'conf', 'z.off']]
display(o7.head(), o7.shape)
```

```python
# Convert seeding to seed strength
# 17 - seed (so 1 seed = strength 16)
# more easily interpreted

df_merged['seed_strength'] = 17 - df_merged['seed_num']
df_merged = df_merged.drop(['seed_num'],axis=1)
```

```python
### Adding conference ranking and converting to conference strength
### Similar to seed strength
### Data only available from 2022 - 2026

max_rank_by_season = ConfRank.groupby(['season','M_W'])['Rank'].max().reset_index()
max_rank_by_season = max_rank_by_season.rename(columns={'Rank': 'max_rank'})

ConfRank_w_max = ConfRank.merge(max_rank_by_season, how='left', on=['season','M_W'])

ConfRank_w_max['conf_strength'] = (ConfRank_w_max['max_rank']+1) - ConfRank_w_max['Rank']

Conf_Strength = ConfRank_w_max.drop(['Rank','Conference','max_rank'],axis=1)

display(Conf_Strength.sort_values(by=['conf','season'],ascending=False).head(), Conf_Strength.shape)
```
