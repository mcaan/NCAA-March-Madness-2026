### Prep the Data

```python
# Create a team x team dataset for predictions
# essentially a dataset of all possible matchups
dyad_tourney = TourneyCompactResults[TourneyCompactResults['Season'].between(2022,2025)]
dyad_tourney.head()

# create dyad
pd.options.mode.chained_assignment = None  # default='warn'

dyad_tourney['team_a'] = np.where(dyad_tourney['WTeamID'] < dyad_tourney['LTeamID'], dyad_tourney['WTeamID'], dyad_tourney['LTeamID'])
dyad_tourney['team_b'] = np.where(dyad_tourney['LTeamID'] < dyad_tourney['WTeamID'], dyad_tourney['WTeamID'], dyad_tourney['LTeamID'])
dyad_tourney['team_a_win'] = np.where(dyad_tourney['team_a'] == dyad_tourney['WTeamID'], 1, 0)

display(dyad_tourney.head(), dyad_tourney.shape)
```

```python
# do the same for the regular season matchups
dyad_regular = RegularCompactResults[RegularCompactResults['Season'].between(2022,2026)]
dyad_regular.head()

# create dyad
pd.options.mode.chained_assignment = None  # default='warn'

dyad_regular['team_a'] = np.where(dyad_regular['WTeamID'] < dyad_regular['LTeamID'], dyad_regular['WTeamID'], dyad_regular['LTeamID'])
dyad_regular['team_b'] = np.where(dyad_regular['LTeamID'] < dyad_regular['WTeamID'], dyad_regular['WTeamID'], dyad_regular['LTeamID'])
dyad_regular['team_a_win'] = np.where(dyad_regular['team_a'] == dyad_regular['WTeamID'], 1, 0)

display(dyad_regular.head(), dyad_regular.shape)
```

```python
# Merge for a full dyad dataset
dyad_full = pd.concat([dyad_tourney,dyad_regular]).reset_index()[['Season','team_a', 'team_b', 'team_a_win']]
display(dyad_full.head(), dyad_full.shape)
```

### Calculating difference in performance between teams

##### The difference in performance statistics will be used rather than incorporating both teams' sets
##### The effect a variable has on increasing team a's odds of winning should have a proportionate effect on decreasing team b's odds
##### Including both, for example, winsbefore_a and winsbefore_b, causes
> ##### Extra variance
> ##### Multicollinearity
> ##### Weaker statistical power
##### Instead, I will include winsbefore_diff = winsbefore_a - winsbefore_b
> ##### This allows me to estimate one coefficient instead of two
##### As a reminder, many metrics are z-scores. So the interpretation for differences becomes relative performance measured in standard deviations of advantage

```python
### Calculate differences
teamxteam2['efg_diff'] = teamxteam2['z.efg_a'] - teamxteam2['z.efg_b']
teamxteam2['sa_diff'] = teamxteam2['z.sa_a'] - teamxteam2['z.sa_b']
teamxteam2['pt_diff'] = teamxteam2['z.pt_a'] - teamxteam2['z.pt_b']
teamxteam2['rank_diff'] = teamxteam2['rank_a'] - teamxteam2['rank_b']
teamxteam2['winsbefore_diff'] = teamxteam2['z.winsbefore_a'] - teamxteam2['z.winsbefore_b']
teamxteam2['winsafter_diff'] = teamxteam2['z.winsafter_a'] - teamxteam2['z.winsafter_b']
teamxteam2['clutch_diff'] = teamxteam2['z.clutch_a'] - teamxteam2['z.clutch_b']
teamxteam2['close_diff'] = teamxteam2['z.close_a'] - teamxteam2['z.close_b']
teamxteam2['grit_diff'] = teamxteam2['z.grit_a'] - teamxteam2['z.grit_b']
teamxteam2['has_ot_diff'] = teamxteam2['z.has_ot_a'] - teamxteam2['z.has_ot_b']
teamxteam2['off_diff'] = teamxteam2['z.off_a'] - teamxteam2['z.off_b']
teamxteam2['def_diff'] = teamxteam2['z.def_a'] - teamxteam2['z.def_b']
teamxteam2['ast_diff'] = teamxteam2['z.ast_a'] - teamxteam2['z.ast_b']
teamxteam2['to_diff'] = teamxteam2['z.to_a'] - teamxteam2['z.to_b']
teamxteam2['stl_diff'] = teamxteam2['z.stl_a'] - teamxteam2['z.stl_b']
teamxteam2['blk_diff'] = teamxteam2['z.blk_a'] - teamxteam2['z.blk_b']
teamxteam2['or_diff'] = teamxteam2['z.or_a'] - teamxteam2['z.or_b']
teamxteam2['dr_diff'] = teamxteam2['z.dr_a'] - teamxteam2['z.dr_b']
teamxteam2['3m_diff'] = teamxteam2['z.3m_a'] - teamxteam2['z.3m_b']
teamxteam2['3a_diff'] = teamxteam2['z.3a_a'] - teamxteam2['z.3a_b']
teamxteam2['winpercintourney_diff'] = teamxteam2['winpercintourney_a'] - teamxteam2['winpercintourney_b']
teamxteam2['seed_diff'] = teamxteam2['seed_strength_a'] - teamxteam2['seed_strength_b']
teamxteam2['conf_diff'] = teamxteam2['conf_strength_a'] - teamxteam2['conf_strength_b']
```
